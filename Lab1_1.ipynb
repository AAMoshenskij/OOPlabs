6. Рассмотрим массив размерности $(6, 7, 8)$. Каков индекс $(i,j,k)$ 100-го элемента?

```python
arr = np.zeros((6, 7, 8))
index_100th_element = np.unravel_index(99, arr.shape) #индексы в Python начинаются с 0, поэтому 100-й элемент имеет индекс 99
print(f"Индекс 100-го элемента: {index_100th_element}")
```

7. Рассмотрим случайную матрицу  $15 × 2 $, представляющую декартовы координаты, преобразуйте их в полярные координаты.

```python
cartesian_coords = np.random.rand(15, 2)
polar_coords = np.zeros_like(cartesian_coords)
polar_coords[:, 0] = np.sqrt(cartesian_coords[:, 0]**2 + cartesian_coords[:, 1]**2)
polar_coords[:, 1] = np.arctan2(cartesian_coords[:, 1], cartesian_coords[:, 0])
print("Декартовы координаты:\n", cartesian_coords)
print("Полярные координаты:\n", polar_coords)
```

8. По двум массивам $x$ и $y$ постройте матрицу Коши $C$
 $(C_{ij} =\frac{1}{x_i - y_j})$

```python
x = np.array([45, 78, 23, 87])
y = np.array([93, 61, 71, 34, 30])
x_matrix = x[:, np.newaxis]
y_matrix = y[np.newaxis, :]
C = 1 / (x_matrix - y_matrix)
print("Матрица Коши C:\n", C)
```

9. Рассмотрим матрицу $10 × 3 $, извлеките строки с неравными значениями (например, $[2,2,3]$).

```python
np.random.seed(0)
matrix = np.random.randint(1, 5, size=(10, 3))
valid_rows_mask = np.apply_along_axis(lambda row: len(np.unique(row)) >= 2 and len(row) != 1, 1, matrix)
valid_rows = matrix[valid_rows_mask]
print("Исходная матрица:\n", matrix)
print("\nСтроки, соответствующие условиям:\n", valid_rows)
```

# Найдите ответы на вопросы с помощью предложенного датасета:

Файл с датасетом - `olympics_dataset.csv`

```python
import pandas as pd
```

```python
# Load dataset here
DATA_PATH = "olympics_dataset.csv"
df = pd.read_csv(DATA_PATH)
df.head()
```

1. Сколько спортсменов участвовали в Летних Олимпийских играх за всё время?

```python
num_athletes = df['player_id'].nunique()
print(num_athletes)
```

2. Найдите участника с наибольшим количеством медалей

```python
#количество медалей для каждого участника
medal_counts = df.loc[df['Medal'] != 'No medal'].groupby('player_id')['Medal'].count().reset_index(name='total_medals')

#участник с максимальным количеством медалей
max_medals = medal_counts['total_medals'].max()
player_with_max_medals = medal_counts.loc[medal_counts['total_medals'] == max_medals, 'player_id'].values[0]

#имя участника с максимальным количеством медалей
player_name = df.loc[df['player_id'] == player_with_max_medals, 'Name'].values[0]

print(f"Участник с наибольшим количеством медалей: {player_name} (id: {player_with_max_medals}, Всего медалей: {max_medals})")
```

3. Какой спорт является самым популярным, то есть в нём участвовало больше всего спортсменов за всё время?

```python
#количество участников в каждом виде спорта
sport_counts = df.groupby('Sport')['player_id'].nunique().reset_index(name='num_participants')

#вид спорта с максимальным количеством участников
most_popular_sport = sport_counts.loc[sport_counts['num_participants'].idxmax(), 'Sport']
max_participants = sport_counts.loc[sport_counts['num_participants'].idxmax(), 'num_participants']

print(f"Самый популярный вид спорта: {most_popular_sport} (Количество участников: {max_participants})")
```

4. Кто был победителем медального зачёта в 2004 году?

```python
# Считаем, что победителем медального зачета является сборная страны, завоевавшая наибольшее число зодотых медалей
# Фильтруем данные только за 2004 год
df_2004 = df[df['Year'] == 2004]

# Считаем количество золотых медалей для каждой страны
gold_medals_by_team = df_2004.loc[df_2004['Medal'] == 'Gold'].groupby('Team')['Medal'].count().reset_index(name='gold_medals')

# Находим страну с максимальным количеством золотых медалей
winning_team = gold_medals_by_team.loc[gold_medals_by_team['gold_medals'].idxmax(), 'Team']
max_gold_medals = gold_medals_by_team.loc[gold_medals_by_team['gold_medals'].idxmax(), 'gold_medals']

print(f"Победитель медального зачета в 2004 году: {winning_team} с {max_gold_medals} золотыми медалями")
```

5. В каком виде спорта США завоевало наибольшее число медалей?

```python
# Фильтруем данные, чтобы оставить только участников из США
usa_df = df[df['Team'] == 'United States']

# Считаем количество медалей для каждого вида спорта, в котором участвовали спортсмены из США
medal_counts_by_sport = usa_df.loc[usa_df['Medal'] != 'No medal'].groupby('Sport')['Medal'].count().reset_index(name='total_medals')

# Находим вид спорта с максимальным количеством медалей
most_successful_sport = medal_counts_by_sport.loc[medal_counts_by_sport['total_medals'].idxmax(), 'Sport']
max_medals = medal_counts_by_sport.loc[medal_counts_by_sport['total_medals'].idxmax(), 'total_medals']

print(f"США завоевало наибольшее число медалей в виде спорта: {most_successful_sport} ({max_medals} медалей)")
```

6. Сколько женщин и мужчин участвовало в Летних Олимпийских играх каждый год?

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Фильтрация только Летних Олимпийских игр
df = df[df['Season'] == 'Summer']

# Группировка по году и полу, подсчет количества участников
gender_counts = df.groupby(['Year', 'Sex'])['player_id'].nunique().reset_index(name='num_participants')

# Создание гистограммы
plt.figure(figsize=(12, 6))
gender_counts.pivot(index='Year', columns='Sex', values='num_participants').plot(kind='bar')
plt.title('Количество женщин и мужчин на Летних Олимпийских играх')
plt.xlabel('Год')
plt.ylabel('Количество участников')
plt.legend(title='Пол')
plt.show()
```

7. Какие 10 стран на текущий момент имеют наибольшее количество золотых медалей?


```python
# Подсчет количества золотых медалей для каждой страны
gold_medals_by_country = df.loc[df['Medal'] == 'Gold'].groupby('Team')['Medal'].count().sort_values(ascending=False).head(10)

# Создание гистограммы
plt.figure(figsize=(12, 6))
gold_medals_by_country.plot(kind='bar')
plt.title('Топ-10 стран по количеству золотых медалей')
plt.xlabel('Страна')
plt.ylabel('Количество золотых медалей')
plt.xticks(rotation=45)
plt.show()
```

8. Сколько золотых, серебряных и бронзовых медалей зарабатывал СССР отдельно по каждым играм?

```python
# Фильтрация записей для СССР
ussr_df = df[df['Team'] == 'Soviet Union']
ussr_df = ussr_df[ussr_df['Medal'] != 'No medal']

# Подсчет количества медалей по типу и году
medal_counts_by_year = ussr_df.groupby(['Year', 'Medal'])['Medal'].count().unstack(fill_value=0)

# Создание гистограммы
plt.figure(figsize=(12, 6))
medal_counts_by_year.plot(kind='bar')
plt.title('Медали СССР на Олимпийских играх')
plt.xlabel('Год')
plt.ylabel('Количество медалей')
plt.legend(title='Тип медали')
plt.show()
```

9. Сколько медалей завоёвывали спортсмены и спортсменки (отдельно) команды Китая на каждой олимпиаде?


```python
china_df = df[df['Team'] == 'China']
medals_df = china_df[china_df['Medal'].isin(['Gold', 'Silver', 'Bronze'])]
medal_counts = medals_df.groupby(['Year', 'Sex']).size().unstack(fill_value=0)
# Настройка графиков
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(10, 10))

# График для мужчин
medal_counts.loc[:, 'M'].plot(kind='bar', ax=axes[0], color='blue')
axes[0].set_title('Количество медалей команды Китая на Олимпиадах (Мужчины)')
axes[0].set_xlabel('Год Олимпиады')
axes[0].set_ylabel('Количество медалей')
axes[0].set_xticks(range(len(medal_counts.index)))
axes[0].set_xticklabels(medal_counts.index, rotation=45)

# График для женщин
medal_counts.loc[:, 'F'].plot(kind='bar', ax=axes[1], color='pink')
axes[1].set_title('Количество медалей команды Китая на Олимпиадах (Женщины)')
axes[1].set_xlabel('Год Олимпиады')
axes[1].set_ylabel('Количество медалей')
axes[1].set_xticks(range(len(medal_counts.index)))
axes[1].set_xticklabels(medal_counts.index, rotation=45)

plt.tight_layout()
plt.show()
```

10. Какие 10 стран имеют наибольшее количество участников, не получивших никаких медалей ни на одной из Летних олимпиад?

```python
# Фильтруем данные, чтобы оставить только участников без медалей
no_medals = df[(df['Season'] == 'Summer') & (df['Medal'] == 'No medal')]
# Считаем количество уникальных участников без медалей для каждой страны
unique_no_medals = no_medals.groupby('Team')['player_id'].nunique()

# Получаем 10 стран с наибольшим количеством уникальных участников без медалей
top_10_countries = unique_no_medals.nlargest(10)

# Построение графика
plt.figure(figsize=(12, 6))
top_10_countries.plot(kind='bar', color='skyblue')
plt.title('Топ 10 стран с наибольшим количеством участников без медалей на Летних олимпиадах')
plt.xlabel('Страны')
plt.ylabel('Количество уникальных участников без медалей')
plt.xticks(rotation=45)
plt.show()
```
